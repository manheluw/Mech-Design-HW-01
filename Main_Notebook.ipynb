(b) Assume the boom has a hollow square cross-section with outer dimensions 2"x2". The material is AISI 1040 steel, quenched and tempered at 650 C (see Shigley Table A-21 in Appendix A). What is the required wall thickness of the square tube to support 2x the rated load?

According to the table sourced above, the yield strength of this material is 95 kpsi. 
Since bending stress is going to dominate here, we will find where the maximum bending stress occurs.

Based off the bending diagram, the maximum moment occurs at the piston location:
![](p3b.png)
$$M_{max} = F \cdot (D_1 - D_2)$$

Using the bending stress equation:

$$\sigma = \frac{M \cdot c}{I}$$

Where:
- $\sigma$ = yield strength = 95 kpsi
- $M$ = maximum bending moment
- $c$ = distance from neutral axis to outer fiber = $\frac{b_{outer}}{2}$ = 1 in
- $I$ = moment of inertia

Solving for the required moment of inertia:

$$I_{required} = \frac{M \cdot c}{\sigma_y}$$

For a hollow square cross-section:

$$I = \frac{b_{outer}^4 - b_{inner}^4}{12}$$

Setting $I = I_{required}$ and solving for $b_{inner}$:

$$b_{inner} = \left( b_{outer}^4 - 12 \cdot I_{required} \right)^{1/4}$$

From $b_{inner}$ we can find the wall thickness:

$$t = \frac{b_{outer} - b_{inner}}{2}$$

```python
b_outer = 2  # in
c = b_outer/2  # in
Stress_max = 95000  # psi
M_max = F * (D1 - D2) * 12  # lb-in (converting ft to in)
I = M_max * c / Stress_max  # in^4
b_inner = (b_outer**4 - 12*I)**0.25  # in
t = (b_outer - b_inner) / 2  # in

print("-" * 45)
print(f"{'M_max':<20} {M_max:>12.1f} {'lb-in':>10}")
print(f"{'I_required':<20} {I:>12.4f} {'in^4':>10}")
print(f"{'b_inner':<20} {b_inner:>12.4f} {'in':>10}")
print(f"{'Wall Thickness (t)':<20} {t:>12.4f} {'in':>10}")
print("-" * 45)
```

## Problem 4: Speed (1994) Elevator Scene Analysis

**What we will determine:** What is supposed to fail first?
- a) The cable
- b) The bolt at the top of the crane
- c) The bolts at the bottom of the crane

---

### Assumptions

**Crane Dimensions:**
- Crane arm: 3m × 3m (isosceles right triangle configuration)
- See FBD below

![Crane FBD](p4crane.png)

**Elevator Properties:**
- Dimensions: 2m × 4m
- Empty weight: ~1000 kg (usually 1000-1500 kg, but this looks like a cheap elevator based on how easily the floor ripped)
- Passengers: ~1000 kg (13 × 80 kg ≈ 1040 kg)
- **Total mass: 2000 kg**

**Fall Analysis:**
- The elevator falls from 2:56 - 3:12, for a total of 16 seconds
- Assuming significant friction from the safety brakes, the effective acceleration is reduced to 4 m/s²
- At 3:12, the elevator stops instantly, then a couple scenes later oscillates about 1/3 m (1 foot), so we'll go with that.
- **Deceleration distance: 0.333 m**

**Cable Properties:**
- Cable angle: 45° from horizontal
- Commercial crane rope with assembly break strength of 51.7 tons
- Reference image of similar cable:

![Cable Reference](p4cable.png)
https://www.uscargocontrol.com/products/1-6x19-galvanized-wire-rope-eips-iwrc-250-coil 


**Bolt Properties:**
- Grade 8 bolts (highest grade SAE bolt)
- Tensile strength: 150,000 psi
- **Shear strength ≈ 60% of tensile strength (for bolts) = 90,000 psi**
- Top bolt: 1.5" diameter (1 bolt)
- Bottom bolts: 1" diameter (4 bolts)

![Grade 8 Bolt Info](p4bolts.png)
https://www.tfgusa.com/high-strength-bolts-for-structural-bolting/

**Force Transfer:**
The cable force acts at 45°, so we must consider how that force transfers to the bolts:
- The top bolt sees shear from the crane arm reaction
- The bottom bolts resist both vertical shear AND the overturning moment from the horizontal force component

---

### Calculations

```python
import numpy as np

# ============================================
# GIVEN / ASSUMPTIONS
# ============================================
m_total = 2000; a_fall = 4; t_fall = 16; d_stop = 1/3; g = 9.81
cable_strength_tons = 51.7; D1 = 3; D2 = 3
tensile_strength_psi = 150000; shear_strength_psi = 0.6 * tensile_strength_psi

# ============================================
# CALCULATIONS
# ============================================
v_impact = a_fall * t_fall
a_decel = v_impact**2 / (2 * d_stop)
F_decel = m_total * a_decel
cable_strength_N = cable_strength_tons * 1000 * g
F_cable_lbf = F_decel * 0.2248
F_vert = F_cable_lbf * np.sin(np.radians(45))
F_horiz = F_cable_lbf * np.cos(np.radians(45))

# Top bolt (1 × 1.5" dia)
A_top = np.pi * 1.5**2 / 4
tau_top = F_vert / A_top

# Bottom bolts (4 × 1" dia)
A_bot = np.pi * 1**2 / 4
M_over = F_horiz * D1 * 39.37
F_tens = M_over / (2 * 12)
sigma_tens = F_tens / A_bot
sigma_vm = np.sqrt(sigma_tens**2 + 3*(F_horiz/(4*A_bot))**2)

# Factors
f_cable = F_decel / cable_strength_N
f_top = tau_top / shear_strength_psi
f_bot = sigma_vm / tensile_strength_psi

# ============================================
# RESULTS
# ============================================
print("RESULTS SUMMARY")
print("-" * 45)
print(f"Impact Velocity:    {v_impact:.1f} m/s")
print(f"Deceleration:       {a_decel/g:.0f} g's")
print(f"Dynamic Force:      {F_decel/1e6:.2f} MN")
print("-" * 45)
print(f"Cable:          {f_cable:>6.1f}x over  --> FAILS")
print(f"Top Bolt:       {f_top:>6.1f}x over  --> FAILS")
print(f"Bottom Bolts:   {f_bot:>6.1f}x over  --> FAILS")
print("-" * 45)
print("FAILURE ORDER: Bottom Bolts > Cable > Top Bolt")
```

### Conclusion

Based on this analysis, **everything fails, no questions asked**. The bottom bolts actually experience the highest stress due to the massive overturning moment from the horizontal cable force acting on a 3m moment arm.

In reality, if the elevator fell for 16 seconds at 4 m/s² acceleration and stopped in only 1/3 of a meter:
1. The cable would snap instantly (24x over capacity)
2. The top bolt would shear off (12x over capacity)  
3. The bottom bolts would rip out of the building (570x over capacity from the moment!)
