```python
# ============================================
# EQUILIBRIUM EQUATIONS - PROBLEMS A, B, C, D
# ============================================

# ============================================
# INPUT PARAMETERS
# ============================================
F1 = 1  # Force 1 [kN]
F2 = 1  # Force 2 [kN]

# ============================================
# PROBLEM (a) - Top View
# Moment arms: 4 m
# ============================================
Rx_a = F1          # ΣFx = 0 = F1 - Rx
Ry_a = 0           # ΣFy = 0
Rz_a = F2          # ΣFz = 0 = Rz - F2
Mx_a = 4 * F2      # ΣMx = 0 = Mx - 4(F2)
My_a = 4 * F2      # ΣMy = 0 = 4(F2) - My
Mz_a = 4 * F1      # ΣMz = 0 = Mz - 4(F1)

# ============================================
# PROBLEM (b) - Front View
# Moment arms: 2 m
# ============================================
Rx_b = 0           # ΣFx = 0
Ry_b = F1          # ΣFy = 0 = Ry - F1
Rz_b = F2          # ΣFz = 0 = Rz - F2
Mx_b = 2 * F1      # ΣMx = 0 = 2(F1) - Mx
My_b = 2 * F2      # ΣMy = 0 = My - 2(F2)
Mz_b = 2 * F1      # ΣMz = 0 = 2(F1) - Mz

# ============================================
# PROBLEM (c) - Front/Right View
# Moment arms: 2 m
# ============================================
Rx_c = 0           # ΣFx = 0
Ry_c = F2          # ΣFy = 0 = Ry - F2
Rz_c = F1          # ΣFz = 0 = F1 - Rz
Mx_c = 2 * F2      # ΣMx = 0 = Mx - 2(F2)
My_c = 2 * F1      # ΣMy = 0 = 2(F1) - My
Mz_c = 0           # ΣMz = 0

# ============================================
# PROBLEM (d) - Top View
# Moment arms: 3 m
# ============================================
Rx_d = 0           # ΣFx = 0
Ry_d = 0           # ΣFy = 0
Rz_d = F1 + F2     # ΣFz = 0 = Rz - F1 - F2
Mx_d = 3 * (F1 + F2)  # ΣMx = 0 = Mx - 3(F1 + F2)
My_d = 0           # ΣMy = 0
Mz_d = 0           # ΣMz = 0

# ============================================
# RESULTS OUTPUT - WIDE TABLE FORMAT
# ============================================
print(f"Given: F1 = {F1} kN, F2 = {F2} kN, Grid = 1 m")
print()

print("Reaction Forces [kN]:")
print("-" * 45)
print(f"{'Problem':<10} {'Rx':>10} {'Ry':>10} {'Rz':>10}")
print("-" * 45)
print(f"{'(a)':<10} {Rx_a:>10.1f} {Ry_a:>10.1f} {Rz_a:>10.1f}")
print(f"{'(b)':<10} {Rx_b:>10.1f} {Ry_b:>10.1f} {Rz_b:>10.1f}")
print(f"{'(c)':<10} {Rx_c:>10.1f} {Ry_c:>10.1f} {Rz_c:>10.1f}")
print(f"{'(d)':<10} {Rx_d:>10.1f} {Ry_d:>10.1f} {Rz_d:>10.1f}")
print()

print("Reaction Moments [kN·m]:")
print("-" * 45)
print(f"{'Problem':<10} {'Mx':>10} {'My':>10} {'Mz':>10}")
print("-" * 45)
print(f"{'(a)':<10} {Mx_a:>10.1f} {My_a:>10.1f} {Mz_a:>10.1f}")
print(f"{'(b)':<10} {Mx_b:>10.1f} {My_b:>10.1f} {Mz_b:>10.1f}")
print(f"{'(c)':<10} {Mx_c:>10.1f} {My_c:>10.1f} {Mz_c:>10.1f}")
print(f"{'(d)':<10} {Mx_d:>10.1f} {My_d:>10.1f} {Mz_d:>10.1f}")
```
